```python

import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt

# c nh t th mc exercise
data = iio.imread('exercise/colorful-ripe-tropical-fruits.jpg')
print("Kch thc nh:", data.shape)

# Ct qu kiwi (ta   xc nh th cng, bn c th iu chnh nu cn)
kiwi = data[706:849, 1420:1571]  # (y1:y2, x1:x2)

# Tnh tin sang phi 30 pixels
height, width, _ = kiwi.shape
translated = np.zeros((height, width + 30, 3), dtype=np.uint8)
translated[:, 30:] = kiwi

# Lu nh  tnh tin
iio.imsave('kiwi_translated.jpg', translated)

# Hin th nh kt qu
plt.imshow(translated)
plt.title("Kiwi  tnh tin +30px")
plt.axis('off')
plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# c nh
image = cv2.imread('exercise/colorful-ripe-tropical-fruits.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Ct vng qu u  v da hu (ta  ng)
papaya = image_rgb[328:806, 239:542]
watermelon = image_rgb[319:784, 1696:1890]


# i mu bng cch o mu
papaya_inv = 255 - papaya
watermelon_inv = 255 - watermelon

# Hin th kt qu
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.title("Original Papaya")
plt.imshow(papaya)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Inverted Papaya")
plt.imshow(papaya_inv)
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Original Watermelon")
plt.imshow(watermelon)
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Inverted Watermelon")
plt.imshow(watermelon_inv)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python

import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# c nh gc
data = iio.imread('exercise/quang_ninh.jpg')
print("Kch thc nh gc:", data.shape)

# ------------------ Ct vng ngn ni v con thuyn ------------------
# Bn c th iu chnh li ta  nu cn
mountain = data[20:320, 429:656]      # ngn ni (v d)
boat = data[463:534, 513:645]       # con thuyn (v d)

# ------------------ Xoay 45  ------------------
mountain_rotated = nd.rotate(mountain, 45, reshape=True)
boat_rotated = nd.rotate(boat, 45, reshape=True)

# ------------------ Lu kt qu ------------------
iio.imsave('mountain_rotated.jpg', mountain_rotated)
iio.imsave('boat_rotated.jpg', boat_rotated)
print("  lu nh: mountain_rotated.jpg v boat_rotated.jpg")

# ------------------ Hin th kt qu ------------------
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(mountain_rotated)
plt.title("Ngn ni xoay 45")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(boat_rotated)
plt.title("Con thuyn xoay 45")
plt.axis('off')

plt.tight_layout()
plt.show()

```

```python

import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# c nh t th mc exercise
data = iio.imread('exercise/pagoda.jpg')
print("Kch thc nh gc:", data.shape)

# Ct vng ngi cha (ta  mu, bn c th iu chnh)
pagoda = data[132:244, 188:380]  # (y1:y2, x1:x2)
print("Kch thc vng cha:", pagoda.shape)

# Phng to 5 ln
pagoda_zoomed = nd.zoom(pagoda, (5, 5, 1))  # Phng to theo (y, x, knh)

# Lu nh kt qu
iio.imsave('pagoda_scaled.jpg', pagoda_zoomed)

# Hin th nh  phng to
plt.imshow(pagoda_zoomed)
plt.title("Ngi cha phng to 5 ln")
plt.axis('off')
plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Danh sch nh mu
image_names = {
    "1": "exercise/pagoda.jpg",
    "2": "exercise/quang_ninh.jpg",
    "3": "exercise/colorful-ripe-tropical-fruits.jpg"
}

# Hm chn nh t menu
def select_image():
    print("\nChn nh:")
    print("1 - pagoda.jpg")
    print("2 - quang_ninh.jpg")
    print("3 - colorful-ripe-tropical-fruits.jpg")
    img_choice = input("Nhp s (1/2/3): ")
    path = image_names.get(img_choice)
    if path:
        image = cv2.imread(path)
        if image is not None:
            return image, path
        else:
            print(" Khng th c nh.")
    else:
        print(" La chn khng hp l.")
    return None, None

# Tnh tin nh
def translate_image(img):
    tx = int(input("Nhp  tnh tin theo trc x: "))
    ty = int(input("Nhp  tnh tin theo trc y: "))
    M = np.float32([[1, 0, tx], [0, 1, ty]])
    return cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

# Xoay nh
def rotate_image(img):
    angle = float(input("Nhp gc xoay (): "))
    h, w = img.shape[:2]
    M = cv2.getRotationMatrix2D((w/2, h/2), angle, 1)
    return cv2.warpAffine(img, M, (w, h))

# Phng to hoc thu nh
def scale_image(img, fx, fy):
    return cv2.resize(img, None, fx=fx, fy=fy, interpolation=cv2.INTER_LINEAR)

# Hin th nh v cho php click ly ta 
def coordinate_map(img):
    clone = img.copy()
    def click_event(event, x, y, flags, param):
        if event == cv2.EVENT_LBUTTONDOWN:
            print(f"Bn  click ti: x = {x}, y = {y}")
            cv2.circle(clone, (x, y), 5, (0, 0, 255), -1)
            cv2.imshow("Coordinate Map", clone)
    cv2.imshow("Coordinate Map", clone)
    cv2.setMouseCallback("Coordinate Map", click_event)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Menu chnh
def main():
    while True:
        print("\n=== MENU ===")
        print("T - Tnh tin")
        print("X - Xoay")
        print("P - Phng to")
        print("H - Thu nh")
        print("C - Coordinate Map")
        print("Q - Thot")
        choice = input("Chn thao tc: ").strip().upper()

        if choice == "Q":
            print("Kt thc chng trnh.")
            break

        image, path = select_image()
        if image is None:
            continue

        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        if choice == "T":
            result = translate_image(image_rgb)
        elif choice == "X":
            result = rotate_image(image_rgb)
        elif choice == "P":
            result = scale_image(image_rgb, fx=2.0, fy=2.0)
        elif choice == "H":
            result = scale_image(image_rgb, fx=0.5, fy=0.5)
        elif choice == "C":
            coordinate_map(image)
            continue
        else:
            print(" La chn khng hp l.")
            continue

        # Hin th kt qu
        plt.figure(figsize=(10, 5))
        plt.imshow(result)
        plt.title(f"Kt qu - {choice}")
        plt.axis('off')
        plt.show()

        # Lu nh kt qu
        save_name = f"output_{choice}.jpg"
        cv2.imwrite(save_name, cv2.cvtColor(result, cv2.COLOR_RGB2BGR))
        print(f"  lu nh kt qu: {save_name}")

# Chy chng trnh
main()
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Danh sch nh mu
image_names = {
    "1": "exercise/pagoda.jpg",
    "2": "exercise/quang_ninh.jpg",
    "3": "exercise/colorful-ripe-tropical-fruits.jpg"
}

# Hm chn nh t menu
def select_image():
    print("\nChn nh:")
    print("1 - pagoda.jpg")
    print("2 - quang_ninh.jpg")
    print("3 - colorful-ripe-tropical-fruits.jpg")
    img_choice = input("Nhp s (1/2/3): ")
    path = image_names.get(img_choice)
    if path:
        image = cv2.imread(path)
        if image is not None:
            return image, path
        else:
            print(" Khng th c nh.")
    else:
        print(" La chn khng hp l.")
    return None, None

# Tnh tin nh
def translate_image(img):
    tx = int(input("Nhp  tnh tin theo trc x: "))
    ty = int(input("Nhp  tnh tin theo trc y: "))
    M = np.float32([[1, 0, tx], [0, 1, ty]])
    return cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

# Xoay nh
def rotate_image(img):
    angle = float(input("Nhp gc xoay (): "))
    h, w = img.shape[:2]
    M = cv2.getRotationMatrix2D((w/2, h/2), angle, 1)
    return cv2.warpAffine(img, M, (w, h))

# Phng to hoc thu nh
def scale_image(img, fx, fy):
    return cv2.resize(img, None, fx=fx, fy=fy, interpolation=cv2.INTER_LINEAR)

# Hin th nh v cho php click ly ta 
def coordinate_map(img):
    clone = img.copy()
    def click_event(event, x, y, flags, param):
        if event == cv2.EVENT_LBUTTONDOWN:
            print(f"Bn  click ti: x = {x}, y = {y}")
            cv2.circle(clone, (x, y), 5, (0, 0, 255), -1)
            cv2.imshow("Coordinate Map", clone)
    cv2.imshow("Coordinate Map", clone)
    cv2.setMouseCallback("Coordinate Map", click_event)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Menu chnh
def main():
    while True:
        print("\n=== MENU ===")
        print("T - Tnh tin")
        print("X - Xoay")
        print("P - Phng to")
        print("H - Thu nh")
        print("C - Coordinate Map")
        print("Q - Thot")
        choice = input("Chn thao tc: ").strip().upper()

        if choice == "Q":
            print("Kt thc chng trnh.")
            break

        image, path = select_image()
        if image is None:
            continue

        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        if choice == "T":
            result = translate_image(image_rgb)
        elif choice == "X":
            result = rotate_image(image_rgb)
        elif choice == "P":
            result = scale_image(image_rgb, fx=2.0, fy=2.0)
        elif choice == "H":
            result = scale_image(image_rgb, fx=0.5, fy=0.5)
        elif choice == "C":
            coordinate_map(image)
            continue
        else:
            print(" La chn khng hp l.")
            continue

        # Hin th kt qu
        plt.figure(figsize=(10, 5))
        plt.imshow(result)
        plt.title(f"Kt qu - {choice}")
        plt.axis('off')
        plt.show()

        # Lu nh kt qu
        save_name = f"output_{choice}.jpg"
        cv2.imwrite(save_name, cv2.cvtColor(result, cv2.COLOR_RGB2BGR))
        print(f"  lu nh kt qu: {save_name}")

# Chy chng trnh
main()
```
