```python

import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt

# Đọc ảnh từ thư mục exercise
data = iio.imread('exercise/colorful-ripe-tropical-fruits.jpg')
print("Kích thước ảnh:", data.shape)

# Cắt quả kiwi (tọa độ đã xác định thủ công, bạn có thể điều chỉnh nếu cần)
kiwi = data[706:849, 1420:1571]  # (y1:y2, x1:x2)

# Tịnh tiến sang phải 30 pixels
height, width, _ = kiwi.shape
translated = np.zeros((height, width + 30, 3), dtype=np.uint8)
translated[:, 30:] = kiwi

# Lưu ảnh đã tịnh tiến
iio.imsave('kiwi_translated.jpg', translated)

# Hiển thị ảnh kết quả
plt.imshow(translated)
plt.title("Kiwi đã tịnh tiến +30px")
plt.axis('off')
plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Đọc ảnh
image = cv2.imread('exercise/colorful-ripe-tropical-fruits.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Cắt vùng quả đu đủ và dưa hấu (tọa độ đúng)
papaya = image_rgb[328:806, 239:542]
watermelon = image_rgb[319:784, 1696:1890]


# Đổi màu bằng cách đảo màu
papaya_inv = 255 - papaya
watermelon_inv = 255 - watermelon

# Hiển thị kết quả
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.title("Original Papaya")
plt.imshow(papaya)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Inverted Papaya")
plt.imshow(papaya_inv)
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Original Watermelon")
plt.imshow(watermelon)
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Inverted Watermelon")
plt.imshow(watermelon_inv)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python

import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# Đọc ảnh gốc
data = iio.imread('exercise/quang_ninh.jpg')
print("Kích thước ảnh gốc:", data.shape)

# ------------------ Cắt vùng ngọn núi và con thuyền ------------------
# Bạn có thể điều chỉnh lại tọa độ nếu cần
mountain = data[20:320, 429:656]      # ngọn núi (ví dụ)
boat = data[463:534, 513:645]       # con thuyền (ví dụ)

# ------------------ Xoay 45 độ ------------------
mountain_rotated = nd.rotate(mountain, 45, reshape=True)
boat_rotated = nd.rotate(boat, 45, reshape=True)

# ------------------ Lưu kết quả ------------------
iio.imsave('mountain_rotated.jpg', mountain_rotated)
iio.imsave('boat_rotated.jpg', boat_rotated)
print("✅ Đã lưu ảnh: mountain_rotated.jpg và boat_rotated.jpg")

# ------------------ Hiển thị kết quả ------------------
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(mountain_rotated)
plt.title("Ngọn núi xoay 45°")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(boat_rotated)
plt.title("Con thuyền xoay 45°")
plt.axis('off')

plt.tight_layout()
plt.show()

```

```python

import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# Đọc ảnh từ thư mục exercise
data = iio.imread('exercise/pagoda.jpg')
print("Kích thước ảnh gốc:", data.shape)

# Cắt vùng ngôi chùa (tọa độ mẫu, bạn có thể điều chỉnh)
pagoda = data[132:244, 188:380]  # (y1:y2, x1:x2)
print("Kích thước vùng chùa:", pagoda.shape)

# Phóng to 5 lần
pagoda_zoomed = nd.zoom(pagoda, (5, 5, 1))  # Phóng to theo (y, x, kênh)

# Lưu ảnh kết quả
iio.imsave('pagoda_scaled.jpg', pagoda_zoomed)

# Hiển thị ảnh đã phóng to
plt.imshow(pagoda_zoomed)
plt.title("Ngôi chùa phóng to 5 lần")
plt.axis('off')
plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Danh sách ảnh mẫu
image_names = {
    "1": "exercise/pagoda.jpg",
    "2": "exercise/quang_ninh.jpg",
    "3": "exercise/colorful-ripe-tropical-fruits.jpg"
}

# Hàm chọn ảnh từ menu
def select_image():
    print("\nChọn ảnh:")
    print("1 - pagoda.jpg")
    print("2 - quang_ninh.jpg")
    print("3 - colorful-ripe-tropical-fruits.jpg")
    img_choice = input("Nhập số (1/2/3): ")
    path = image_names.get(img_choice)
    if path:
        image = cv2.imread(path)
        if image is not None:
            return image, path
        else:
            print("❌ Không thể đọc ảnh.")
    else:
        print("❌ Lựa chọn không hợp lệ.")
    return None, None

# Tịnh tiến ảnh
def translate_image(img):
    tx = int(input("Nhập độ tịnh tiến theo trục x: "))
    ty = int(input("Nhập độ tịnh tiến theo trục y: "))
    M = np.float32([[1, 0, tx], [0, 1, ty]])
    return cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

# Xoay ảnh
def rotate_image(img):
    angle = float(input("Nhập góc xoay (độ): "))
    h, w = img.shape[:2]
    M = cv2.getRotationMatrix2D((w/2, h/2), angle, 1)
    return cv2.warpAffine(img, M, (w, h))

# Phóng to hoặc thu nhỏ
def scale_image(img, fx, fy):
    return cv2.resize(img, None, fx=fx, fy=fy, interpolation=cv2.INTER_LINEAR)

# Hiển thị ảnh và cho phép click lấy tọa độ
def coordinate_map(img):
    clone = img.copy()
    def click_event(event, x, y, flags, param):
        if event == cv2.EVENT_LBUTTONDOWN:
            print(f"Bạn đã click tại: x = {x}, y = {y}")
            cv2.circle(clone, (x, y), 5, (0, 0, 255), -1)
            cv2.imshow("Coordinate Map", clone)
    cv2.imshow("Coordinate Map", clone)
    cv2.setMouseCallback("Coordinate Map", click_event)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Menu chính
def main():
    while True:
        print("\n=== MENU ===")
        print("T - Tịnh tiến")
        print("X - Xoay")
        print("P - Phóng to")
        print("H - Thu nhỏ")
        print("C - Coordinate Map")
        print("Q - Thoát")
        choice = input("Chọn thao tác: ").strip().upper()

        if choice == "Q":
            print("Kết thúc chương trình.")
            break

        image, path = select_image()
        if image is None:
            continue

        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        if choice == "T":
            result = translate_image(image_rgb)
        elif choice == "X":
            result = rotate_image(image_rgb)
        elif choice == "P":
            result = scale_image(image_rgb, fx=2.0, fy=2.0)
        elif choice == "H":
            result = scale_image(image_rgb, fx=0.5, fy=0.5)
        elif choice == "C":
            coordinate_map(image)
            continue
        else:
            print("❌ Lựa chọn không hợp lệ.")
            continue

        # Hiển thị kết quả
        plt.figure(figsize=(10, 5))
        plt.imshow(result)
        plt.title(f"Kết quả - {choice}")
        plt.axis('off')
        plt.show()

        # Lưu ảnh kết quả
        save_name = f"output_{choice}.jpg"
        cv2.imwrite(save_name, cv2.cvtColor(result, cv2.COLOR_RGB2BGR))
        print(f"✅ Đã lưu ảnh kết quả: {save_name}")

# Chạy chương trình
main()
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Danh sách ảnh mẫu
image_names = {
    "1": "exercise/pagoda.jpg",
    "2": "exercise/quang_ninh.jpg",
    "3": "exercise/colorful-ripe-tropical-fruits.jpg"
}

# Hàm chọn ảnh từ menu
def select_image():
    print("\nChọn ảnh:")
    print("1 - pagoda.jpg")
    print("2 - quang_ninh.jpg")
    print("3 - colorful-ripe-tropical-fruits.jpg")
    img_choice = input("Nhập số (1/2/3): ")
    path = image_names.get(img_choice)
    if path:
        image = cv2.imread(path)
        if image is not None:
            return image, path
        else:
            print("❌ Không thể đọc ảnh.")
    else:
        print("❌ Lựa chọn không hợp lệ.")
    return None, None

# Tịnh tiến ảnh
def translate_image(img):
    tx = int(input("Nhập độ tịnh tiến theo trục x: "))
    ty = int(input("Nhập độ tịnh tiến theo trục y: "))
    M = np.float32([[1, 0, tx], [0, 1, ty]])
    return cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

# Xoay ảnh
def rotate_image(img):
    angle = float(input("Nhập góc xoay (độ): "))
    h, w = img.shape[:2]
    M = cv2.getRotationMatrix2D((w/2, h/2), angle, 1)
    return cv2.warpAffine(img, M, (w, h))

# Phóng to hoặc thu nhỏ
def scale_image(img, fx, fy):
    return cv2.resize(img, None, fx=fx, fy=fy, interpolation=cv2.INTER_LINEAR)

# Hiển thị ảnh và cho phép click lấy tọa độ
def coordinate_map(img):
    clone = img.copy()
    def click_event(event, x, y, flags, param):
        if event == cv2.EVENT_LBUTTONDOWN:
            print(f"Bạn đã click tại: x = {x}, y = {y}")
            cv2.circle(clone, (x, y), 5, (0, 0, 255), -1)
            cv2.imshow("Coordinate Map", clone)
    cv2.imshow("Coordinate Map", clone)
    cv2.setMouseCallback("Coordinate Map", click_event)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Menu chính
def main():
    while True:
        print("\n=== MENU ===")
        print("T - Tịnh tiến")
        print("X - Xoay")
        print("P - Phóng to")
        print("H - Thu nhỏ")
        print("C - Coordinate Map")
        print("Q - Thoát")
        choice = input("Chọn thao tác: ").strip().upper()

        if choice == "Q":
            print("Kết thúc chương trình.")
            break

        image, path = select_image()
        if image is None:
            continue

        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        if choice == "T":
            result = translate_image(image_rgb)
        elif choice == "X":
            result = rotate_image(image_rgb)
        elif choice == "P":
            result = scale_image(image_rgb, fx=2.0, fy=2.0)
        elif choice == "H":
            result = scale_image(image_rgb, fx=0.5, fy=0.5)
        elif choice == "C":
            coordinate_map(image)
            continue
        else:
            print("❌ Lựa chọn không hợp lệ.")
            continue

        # Hiển thị kết quả
        plt.figure(figsize=(10, 5))
        plt.imshow(result)
        plt.title(f"Kết quả - {choice}")
        plt.axis('off')
        plt.show()

        # Lưu ảnh kết quả
        save_name = f"output_{choice}.jpg"
        cv2.imwrite(save_name, cv2.cvtColor(result, cv2.COLOR_RGB2BGR))
        print(f"✅ Đã lưu ảnh kết quả: {save_name}")

# Chạy chương trình
main()
```
