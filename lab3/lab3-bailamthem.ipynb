```python
#2.Tịnh tiến ảnh
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('fruit.jpg', mode='L')
bdata = nd.shift(data, (50, -30))
plt.imshow(bdata, cmap='gray')
plt.show()

# nhận xét
# sau khi thay đổi (100, 25) thành (50, -30) thì +50 hàng → ảnh bị dịch xuống dưới

#  -30 cột → ảnh bị dịch sang trái
```

```python
# 3. thay đổi kích thước ảnh
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# Đọc ảnh
data = iio.imread('fruit.jpg')
print("Kích thước ảnh gốc:", data.shape)

# ----------- Phóng to gấp 3 lần -----------
zoomed_up = nd.zoom(data, (3, 3, 1))  # (tỷ lệ theo y, x, kênh)
print("Phóng to 3x:", zoomed_up.shape)

# ----------- Thu nhỏ còn 0.3 lần -----------
zoomed_down = nd.zoom(data, (0.3, 0.3, 1))
print("Thu nhỏ 0.3x:", zoomed_down.shape)

# ----------- Hiển thị kết quả thu nhỏ -----------
plt.imshow(zoomed_down)
plt.title("Ảnh thu nhỏ 0.3 lần")
plt.axis('off')
plt.show()


```

```python
#4 Xoay ảnh
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# Đọc ảnh
data = iio.imread('fruit.jpg')
print("Kích thước ảnh gốc:", data.shape)

# ----------- Xoay 45 độ với reshape=True (mặc định) -----------
rotated1 = nd.rotate(data, 45)  # reshape=True là mặc định
print("Kích thước ảnh xoay (reshape=True):", rotated1.shape)

plt.imshow(rotated1)
plt.title("Xoay 45° - reshape=True")
plt.axis('off')
plt.show()

# ----------- Xoay 45 độ với reshape=False -----------
rotated2 = nd.rotate(data, 45, reshape=False)
print("Kích thước ảnh xoay (reshape=False):", rotated2.shape)

plt.imshow(rotated2)
plt.title("Xoay 45° - reshape=False")
plt.axis('off')
plt.show()
```
