```python
#2.Tnh tin nh
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('fruit.jpg', mode='L')
bdata = nd.shift(data, (50, -30))
plt.imshow(bdata, cmap='gray')
plt.show()

# nhn xt
# sau khi thay i (100, 25) thnh (50, -30) th +50 hng  nh b dch xung di

#  -30 ct  nh b dch sang tri
```

```python
# 3. thay i kch thc nh
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# c nh
data = iio.imread('fruit.jpg')
print("Kch thc nh gc:", data.shape)

# ----------- Phng to gp 3 ln -----------
zoomed_up = nd.zoom(data, (3, 3, 1))  # (t l theo y, x, knh)
print("Phng to 3x:", zoomed_up.shape)

# ----------- Thu nh cn 0.3 ln -----------
zoomed_down = nd.zoom(data, (0.3, 0.3, 1))
print("Thu nh 0.3x:", zoomed_down.shape)

# ----------- Hin th kt qu thu nh -----------
plt.imshow(zoomed_down)
plt.title("nh thu nh 0.3 ln")
plt.axis('off')
plt.show()


```

```python
#4 Xoay nh
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# c nh
data = iio.imread('fruit.jpg')
print("Kch thc nh gc:", data.shape)

# ----------- Xoay 45  vi reshape=True (mc nh) -----------
rotated1 = nd.rotate(data, 45)  # reshape=True l mc nh
print("Kch thc nh xoay (reshape=True):", rotated1.shape)

plt.imshow(rotated1)
plt.title("Xoay 45 - reshape=True")
plt.axis('off')
plt.show()

# ----------- Xoay 45  vi reshape=False -----------
rotated2 = nd.rotate(data, 45, reshape=False)
print("Kch thc nh xoay (reshape=False):", rotated2.shape)

plt.imshow(rotated2)
plt.title("Xoay 45 - reshape=False")
plt.axis('off')
plt.show()
```
